##Para la Media

Un intervalo de confianza del 99% para el tiempo promedio de producción por unidad de cada una de las líneas. Compare y analice. Explique para qué le podría servir la construcción de estos intervalos.

 Para la Linea 1

```python
from scipy.stats import norm, chi2, f

media_linea1 = 24.877143
desviacion_linea1 = 1.974118
varianza_linea1 = 3.897140
n1 = 98


z = norm.ppf(0.995, 0,1)   # Por izquierda
print(z)

#Tambien podemos obtener el valor critico a mano y simplemente declarar la variables



#Intervalos por derecha e izquierda
ic_media_linea1 = (media_linea1 - z * (desviacion_linea1 / np.sqrt(n1)), media_linea1 + z * (desviacion_linea1 / np.sqrt(n1)))

print("Intervalo de confianza para la media de Línea 1:", ic_media_linea1)
```

```python
x = np.linspace(24, 26)

pdf = norm.pdf(x, media_linea1, desviacion_linea1 / np.sqrt(n1))  #Acumulada

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal', color='blue')
plt.fill_between(x, pdf, where=(x >= ic_media_linea1[0]) & (x <= ic_media_linea1[1]),
                 color='blue', alpha=0.3, label='Intervalo de Confianza 99%')
plt.axvline(ic_media_linea1[0], color='red', linestyle='--', label='Límite Izquierdo')
plt.axvline(ic_media_linea1[1], color='green', linestyle='--', label='Límite Derecho')
plt.axvline(media_linea1, color='black', linestyle='-', label='Media')
plt.title('Intervalo de Confianza para la Media de Línea 1')
plt.xlabel('Valores de X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid()
plt.show()
```

## Para la desviación

Un intervalo de confianza del 99% para la desviación del tiempo de producción por unidad de cada una de las líneas de producción. Compare y analice. Explique para qué le podría servir la construcción de estos intervalos.

Para la linea 1

```python
n1 = 98
varianza_linea1 = 3.897140

chi2_inferior = chi2.ppf(0.005, df=n1-1)
print(f"Valor crítico inferior: {chi2_inferior}")
chi2_superior = chi2.ppf(0.995, df=n1-1)
print(f"Valor crítico superior: {chi2_superior}")


ic_desviacion_linea1 = (np.sqrt((n1 - 1) * varianza_linea1 / chi2_superior),
                        np.sqrt((n1 - 1) * varianza_linea1 / chi2_inferior))

print("Intervalo de confianza para la desviación estándar de Línea 1:", ic_desviacion_linea1)
```

```python

desviaciones = np.linspace(1.4, 2.8, 1000)
chi2_values = chi2.pdf((n1 - 1) * varianza_linea1 / desviaciones**2, df=n1 - 1)

# Graficar la densidad de probabilidad de chi cuadrado y los límites del intervalo de confianza
plt.figure(figsize=(10, 6))
plt.plot(desviaciones, chi2_values, label='Densidad Chi Cuadrado', color='b')
plt.fill_between(desviaciones, chi2_values, where=(desviaciones >= ic_desviacion_linea1[0]) &
                 (desviaciones <= ic_desviacion_linea1[1]), color='lightgreen', alpha=0.5, label='Intervalo de Confianza')
plt.axvline(ic_desviacion_linea1[0], color='red', linestyle='--', label=f'Limite Inferior: {ic_desviacion_linea1[0]:.3f}')
plt.axvline(ic_desviacion_linea1[1], color='green', linestyle='--', label=f'Limite Superior: {ic_desviacion_linea1[1]:.3f}')

# Etiquetas y leyenda
plt.xlabel('Desviación estándar')
plt.ylabel('Densidad de probabilidad')
plt.title('Intervalo de confianza para la desviación estándar de Línea 1')
plt.legend()
plt.grid(True)
plt.show()
```

Para la linea 2

```python
n2 = 105
varianza_linea2 = 0.391838

chi2_inferior = chi2.ppf(0.005, df=n2-1)
print(f"Valor crítico inferior: {chi2_inferior}")
chi2_superior = chi2.ppf(0.995, df=n2-1)
print(f"Valor crítico superior: {chi2_superior}")


ic_desviacion_linea2 = (np.sqrt((n2 - 1) * varianza_linea2 / chi2_superior),
                        np.sqrt((n2 - 1) * varianza_linea2 / chi2_inferior))

print("Intervalo de confianza para la desviación estándar de Línea 2:", ic_desviacion_linea2)
```

```python
desviaciones = np.linspace(0.4, 0.85, 1000)
chi2_values = chi2.pdf((n1 - 1) * varianza_linea2 / desviaciones**2, df=n1 - 1)

# Graficar la densidad de probabilidad de chi cuadrado y los límites del intervalo de confianza
plt.figure(figsize=(10, 6))
plt.plot(desviaciones, chi2_values, label='Densidad Chi Cuadrado', color='b')
plt.fill_between(desviaciones, chi2_values, where=(desviaciones >= ic_desviacion_linea2[0]) &
                 (desviaciones <= ic_desviacion_linea2[1]), color='peachpuff', alpha=0.5, label='Intervalo de Confianza')
plt.axvline(ic_desviacion_linea2[0], color='red', linestyle='--', label=f'Limite Inferior: {ic_desviacion_linea2[0]:.3f}')
plt.axvline(ic_desviacion_linea2[1], color='green', linestyle='--', label=f'Limite Superior: {ic_desviacion_linea2[1]:.3f}')

# Etiquetas y leyenda
plt.xlabel('Desviación estándar')
plt.ylabel('Densidad de probabilidad')
plt.title('Intervalo de confianza para la desviación estándar de Línea 2')
plt.legend()
plt.grid(True)
plt.show()
```

## Para la relación de varianzas

Un intervalo de confianza del 99% para la desviación del tiempo de producción por unidad de cada una de las líneas de producción. Compare y analice. Explique para qué le podría servir la construcción de estos intervalos.

```python
n1=98
n2=105
varianza_linea1 = 3.897140
varianza_linea2 = 0.391838

#Aqui en la distribución F siempre la da en cola izquierda, para que este en cola derecha: 1-0.005

f_inferior = f.ppf(0.995, dfn=n1-1, dfd=n2-1)
print(f"Valor critico inferior: {f_inferior}")
f_superior = f.ppf(0.995, dfn=n2-1, dfd=n1-1)
print(f"Valor critico superior: {f_superior}")
ic_relacion_varianzas = ((varianza_linea1 / varianza_linea2) * (1 / f_inferior),(varianza_linea1 / varianza_linea2) * (f_superior))

print("Intervalo de confianza para la relación de varianzas:", ic_relacion_varianzas)
```

#**Punto 2**

Una máquina que expende bebidas gaseosas está calibrada de modo que descargue un promedio de 𝜇 mililitros por vaso. Si la cantidad de líquido está distribuida normalmente con una desviación estándar igual a 𝜎 ml. Tanto 𝜇 como 𝜎 son desconocidos, pero se cuenta con una muestra de doscientas observaciones para estimarlos. Utilice los estadísticos 𝑋 y 𝑆 como estimadores de 𝜇 y 𝜎, respectivamente, para responder los siguientes literales. Recuerde que:

$$ \bar{X} = \frac{1}{n} \sum_{i=1}^{n}x_{i} $$

$$ S^{2}= \frac{1}{n-1}\sum_{i=1}^{n} (x_{i}-\bar{x})$$

Vamos a leer el archivo que contiene los datos

```python
datos_bebidas= pd.read_excel('/content/datos bebidas. (1).xlsx')
df = pd.DataFrame(datos_bebidas)
df.columns = ['Registro', 'Volumen']

df
```

```python

# Calcular la media muestral (X barra)
media_muestral = df['Volumen'].mean()

# Calcular la varianza muestral (S^2)
varianza_muestral = df['Volumen'].var()

# Calcular desviación muestral (S)
desviacion_muestral = df['Volumen'].std()



print("Media muestral (X̄):", media_muestral)
print("Varianza muestral (S^2):", varianza_muestral)
print("Desviación muestral (S):", desviacion_muestral)

```

```python
#Por izquierda
izquierda= norm.cdf(153, media_muestral, desviacion_muestral)
print(f"P(x<153) = {izquierda}")

#Por derecha
derecha= 1 - norm.cdf(153, media_muestral, desviacion_muestral)
print(f"P(x>153) = {derecha}")

#Por dos colas
dos_colas= 2 * (1 - norm.cdf(153, media_muestral, desviacion_muestral))
print(f"Por dos colas = {dos_colas}")
```

 a. ¿Qué porcentaje de vasos contendrá menos de 153 ml?  

$$P(X<153)$$

```python
#Por izquierda
p = norm.cdf(153, media_muestral, desviacion_muestral)
print(f"P(X < 153) = {p}")
```

b. ¿Cuál es la probabilidad de que un vaso contenga entre 153 y 157 ml?

$$P(153<X<157)$$

Usamos la propiedad:

$$F(b) - F(a) = P(X<b) - P(X<a)$$

---

$$ P(X<157) - P(X<153)$$

```python
x1 = norm.cdf(157, media_muestral, desviacion_muestral)
x2 = norm.cdf(153, media_muestral, desviacion_muestral)
resultado = x1-x2
print(f"P(153<X<157) = {resultado}")
```

c.  Si se usan vasos de 160 ml, ¿cuántos de los siguientes 850 vasos se derramarán?


$$P(X>160)$$

```python
1 - norm.cdf(160, media_muestral, desviacion_muestral)
print(f"P(X > 160) = {resultado}")

derramar = 850 * (1 - norm.cdf(160, media_muestral, desviacion_muestral))
print(f"Vasos que se derramarán: {derramar}")
```

d.  ¿Bajo qué valor estará el 5% de los vasos con menos contenido?

$$P(X < x) = 0.05$$

```python
valor_x =norm.ppf(0.05, media_muestral, desviacion_muestral)
print(f"valor de x cuando la probabilidad es de 0.05: {valor_x}")
```
